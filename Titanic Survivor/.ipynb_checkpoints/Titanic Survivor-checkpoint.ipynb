```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import math
import numpy as np
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
train.columns
```

# Price Class Feature

```python
features = ['Pclass']
X = train[features]
y = train.Survived

train_X, val_X, train_y, val_y = train_test_split(X, y)
```

```python
# DecisionTree Model
fare_model_dt = DecisionTreeRegressor()
fare_model_dt.fit(train_X, train_y)
fare_model_dt_predict = fare_model_dt.predict(val_X)

mae_dt = mean_absolute_error(val_y, fare_model_dt_predict)

mae_dt
```

```python
# RandomForest Model
fare_model_rf = RandomForestRegressor()
fare_model_rf.fit(train_X, train_y)
fare_model_rf_predict = fare_model_rf.predict(val_X)

mae_rf = mean_absolute_error(val_y, fare_model_rf_predict)

mae_rf
```

```python
# LinearRegression Model
fare_model_lr = LinearRegression()
fare_model_lr.fit(train_X, train_y)
fare_model_lr_predict = fare_model_lr.predict(val_X)

mae_lr = mean_absolute_error(val_y, fare_model_lr_predict)

mae_lr
```

```python
mae_values_pclass = [mae_rf, mae_dt, mae_lr]
mae_models = ['RandomForest', 'DecisionTree', 'LinearRegression']
min_mae_pclass = math.inf
best_model_pclass = 'a'
for i, j in zip(mae_values_pclass, mae_models):
    print ('Mean Absolute Error from %s Model is: %.5f' % (j, i))
    if i < min_mae_pclass:
        min_mae_pclass = i
        best_model_pclass = j     
print ('\nBest model based on Class Fare is %s Model with a Mean Absolute Error equals to %.5f' %(best_model_pclass, min_mae_pclass))
```

```python
survivor_count = train.groupby(['Survived', 'Pclass']).PassengerId.count().reset_index()
survivor_count
```

```python
survivor_count_pivot = survivor_count.pivot(columns='Survived', index='Pclass', values='PassengerId').reset_index()
survivor_count_pivot.columns =['Pclass', 'Not_Survived', 'Survived']
survivor_count_pivot
```

```python
survivor_count_pivot['Total'] = survivor_count_pivot.Not_Survived + survivor_count_pivot.Survived
survivor_count_pivot['Percentage Survived'] = survivor_count_pivot.Survived / survivor_count_pivot.Total * 1.0
survivor_count_pivot
```

```python
plt.bar(survivor_count_pivot['Pclass'], survivor_count_pivot['Percentage Survived'])
ax = plt.subplot()
ax.set_xticks(range(1, 4))
ax.set_xlabel('Fare Class')
ax.set_ylabel('Percent Survived')
ax.set_title('Titanic Survived by Fare Clcass')
plt.show()
```

# Gender Feature

```python
train['Sex'] = train.Sex.apply(lambda x: 1 if x == 'male' else 0)
gender_feature = ['Sex']
X_gender = train[gender_feature]
train_X_g, val_X_g, train_y_g, val_y_g = train_test_split(X_gender, y)
```

```python
# RandomForest Regressor Model
gender_model_rf = RandomForestRegressor(random_state = 6)
gender_model_rf.fit(train_X_g, train_y_g)
gender_model_rf_predict = gender_model_rf.predict(val_X_g)
mae_gender_rf = mean_absolute_error(val_y_g, gender_model_rf_predict)

mae_gender_rf
```

```python
# DecisionTreeRegressor Model
gender_model_dt = DecisionTreeRegressor()
gender_model_dt.fit(train_X_g, train_y_g)
gender_model_dt_predict = gender_model_dt.predict(val_X_g)
mae_gender_dt = mean_absolute_error(val_y_g, gender_model_dt_predict)

mae_gender_dt
```

```python
# LinearRegression Model
gender_model_lr = LinearRegression()
gender_model_lr.fit(train_X_g, train_y_g)
gender_model_lr_predict = gender_model_lr.predict(val_X_g)
mae_gender_lr = mean_absolute_error(val_y_g, gender_model_lr_predict)

mae_gender_lr
```

```python
mae_values_gender = [mae_gender_rf, mae_gender_dt, mae_gender_lr]
min_mae_gender = math.inf
best_model_gender = 'a'
for i, j in zip(mae_values_gender, mae_models):
    print ('Mean Absolute Error from %s Model is: %.5f' % (j ,i))
    if i < min_mae_gender:
        min_mae_gender = i
        best_model_gender = j
print ('\nBest model based on Gender is %s Model with a Mean Absolute Error equals to %.5f' %(best_model_gender, min_mae_gender))
```

# Combined Features: Gender and Fare Class

```python
combined_features = ['Pclass', 'Sex']
X_combined = train[combined_features]
train_X_combined, val_X_combined, train_y_combined, val_y_combined = train_test_split(X_combined, y)
```

```python
# RandomForest Regressor Model
combined_rf_model = RandomForestRegressor()
combined_rf_model.fit(train_X_combined, train_y_combined)
combined_rf_predict = combined_rf_model.predict(val_X_combined)
mae_combined_rf = mean_absolute_error(val_y_combined, combined_rf_predict)

mae_combined_rf
```

```python
# DecisionTree Regressor
combined_dt_model = DecisionTreeRegressor()
combined_dt_model.fit(train_X_combined, train_y_combined)
combined_dt_predict = combined_dt_model.predict(val_X_combined)
mae_combined_dt = mean_absolute_error(val_y_combined, combined_dt_predict)

mae_combined_dt
```

```python
# LinearRegression Model
combined_lr_model = LinearRegression()
combined_lr_model.fit(train_X_combined, train_y_combined)
combined_lr_predict = combined_lr_model.predict(val_X_combined)
mae_combined_lr = mean_absolute_error(val_y_combined, combined_lr_predict)

mae_combined_lr
```

```python
mae_values_combined = [mae_combined_rf, mae_combined_dt, mae_combined_lr]
min_mae_combined = math.inf
best_model_combined = 'a'
for i, j in zip(mae_values_combined, mae_models):
    print ('Mean Absolute Error from %s Model is: %.5f' % (j ,i))
    if i < min_mae_combined:
        min_mae_combined = i
        best_model_combined = j
print ('\nBest model based on both Gender and Class Fare is %s Model with a Mean Absolute Error equals to %.5f' %(best_model_combined, min_mae_combined))
```
